```python
import re

def clean(txt):
  txt = txt.lower()
  txt = re.sub(r'@[A-Za-z0-9]+', '',txt)
  txt = re.sub(r'#', '',txt)
  txt = re.sub(r'RT[\s]+','',txt)
  txt = re.sub(r'https?:\/\/\S+', '',txt)
# Remove special characters using regex
  txt = re.sub('[^A-Za-z0-9 ]+', '', txt)

  return txt

tweetdata['tweets']=tweetdata['content'].apply(clean)

tweetdata
```

```python
from textblob import TextBlob

# Perform sentiment analysis on each tweet
sentiment_list = []
for tweet in tweets_df['tweets']:
    blob = TextBlob(tweet)
    sentiment = blob.sentiment.polarity
    if sentiment > 0:
        sentiment_list.append('Positive')
    elif sentiment < 0:
        sentiment_list.append('Negative')
    else:
        sentiment_list.append('Neutral')

# Add sentiment column to DataFrame
tweets_df['sentiment'] = sentiment_list
sentiment_list
```

```python
import pandas as pd
import folium
from geopy.geocoders import Nominatim

# Read the CSV file containing the extracted tweets data
#tweets_df = pd.read_csv('tweets_data.csv')

# Group tweets by state or city and sentiment, and count the number of tweets in each group
counts = tweets_df.groupby(['state_or_city', 'sentiment']).size().reset_index(name='count')

# Pivot the counts dataframe to create a pivot table with state or city as the index, sentiment as the columns, and count as the values
pivot_table = counts.pivot(index='state_or_city', columns='sentiment', values='count')

# Normalize the pivot table by dividing each value by the sum of values for each state or city
pivot_table_norm = pivot_table.div(pivot_table.sum(axis=1), axis=0)

# Use geocode to get the latitude and longitude coordinates for each state or city

geolocator = Nominatim(user_agent='my-app', timeout=10)
locations = []
for state_or_city in pivot_table_norm.index:
    location = geolocator.geocode(state_or_city)
    if location is not None:
        locations.append((state_or_city, location.latitude, location.longitude))

# Create a folium map centered on the United States
world_map = folium.Map(location=[0, 0], zoom_start=2)

```
